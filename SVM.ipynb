```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
import cv2
import os
```

```python
#DATA COLLECTION AND ANALYSIS
#loading datasets
dataset=pd.read_csv('/content/train.csv')
```

```python
#separating data and labels
x=dataset.drop(columns='category',axis=1)
y=dataset['category']
print(x)
```

```python
print(y)
```

```python
flatt = dataset.values.flatten()

# Keep only numeric values
numeric_flatt = pd.to_numeric(flatt, errors='coerce')  # non-numeric values become NaN
numeric_flatt = numeric_flatt[~np.isnan(numeric_flatt)]  # remove NaNs

print(np.std(numeric_flatt))
```

```python
#std is so highhh
#applying stndard deviation
scaler=StandardScaler()
```

```python
numeric_data = dataset.select_dtypes(include=['number'])
scaler = StandardScaler()
standard_data = scaler.fit(numeric_data)
final_standard = scaler.transform(numeric_data)

print(final_standard)
```

```python
print(np.std(final_standard))
```

```python
#now training and testing part
#training will be applied on the dataset except outcome
x=final_standard  #standardised data
y=dataset['category']
```

```python
print(dataset['category'].value_counts())
```

```python
# For example, Sentiment140 uses 0 for negative, 4 for positive
dataset['category'] = dataset['category'].replace(4, 1)  # Convert 4 → 1 if needed
print(dataset['category'].value_counts())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#training the dataset
classifier=svm.SVC(kernel='linear')   #svc=support vector classification
#kernel decides how the input data is transformed before classification
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.fit_transform(x_test)
```

```python
classifier.fit(x_train_scaled, y_train)
```

```python
classifier.fit(x_test_scaled, y_test)
```

```python
#model evaluate
#accuracy score
x_train_pred=classifier.predict(x_train_scaled)  #used to make prediction
train_acc=accuracy_score(y_train,x_train_pred)
print(train_acc)
```

```python
#we have to apply the accuracy_score on test data
x_test_pred=classifier.predict(x_test_scaled)
test_acc=accuracy_score(y_test,x_test_pred)
print(test_acc)
```
